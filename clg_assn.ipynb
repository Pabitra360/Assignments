```python
#Q5:WAP to merge two already sorted list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node

    def merge_sorted(self, other):
        dummy = Node(0)
        tail = dummy

        l1, l2 = self.head, other.head
        while l1 and l2:
            if l1.data <= l2.data:
                tail.next = l1
                l1 = l1.next
            else:
                tail.next = l2
                l2 = l2.next
            tail = tail.next

        tail.next = l1 if l1 else l2
        return dummy.next

    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("NULL")

if __name__ == "__main__":
    list1 = LinkedList()
    list2 = LinkedList()

    # Adding nodes to the first sorted list
    list1.insert_end(1)
    list1.insert_end(3)
    list1.insert_end(5)

    # Adding nodes to the second sorted list
    list2.insert_end(2)
    list2.insert_end(4)
    list2.insert_end(6)

    print("List 1: ", end="")
    list1.print_list()
    print("List 2: ", end="")
    list2.print_list()

    merged_head = list1.merge_sorted(list2)
    merged_list = LinkedList()
    merged_list.head = merged_head
    print("Merged List: ", end="")
    merged_list.print_list()
```

```python
# Tower of hanoi
def tower_of_hanoi(n, from_peg, to_peg, aux_peg):
    if n == 1:
        print(f"Move disk 1 from peg {from_peg} to peg {to_peg}")
        return
    tower_of_hanoi(n - 1, from_peg, aux_peg, to_peg)
    print(f"Move disk {n} from peg {from_peg} to peg {to_peg}")
    tower_of_hanoi(n - 1, aux_peg, to_peg, from_peg)

if __name__ == "__main__":
    n = 4  # Number of disks
    print(f"Tower of Hanoi solution for {n} disks:")
    tower_of_hanoi(n, 'A', 'C', 'B')  # A, B and C are names of the pegs
```

```python
#1. WAP to implement a singly circular linkedlist
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.last = None

    def add(self, data):
        new_node = Node(data)
        if not self.last:
            self.last = new_node
            self.last.next = self.last
        else:
            new_node.next = self.last.next
            self.last.next = new_node
            self.last = new_node

    def display(self):
        if not self.last:
            return
        current = self.last.next
        while True:
            print(current.data, end=" ")
            current = current.next
            if current == self.last.next:
                break
        print()

# Example usage
cll = CircularLinkedList()
cll.add(1)
cll.add(2)
cll.add(3)
cll.display()
```

```python
#7. WAP to remove a particular character from a doubly linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node
            new_node.prev = temp

    def remove_character(self, char):
        temp = self.head
        while temp:
            if temp.data == char:
                if temp.prev:
                    temp.prev.next = temp.next
                if temp.next:
                    temp.next.prev = temp.prev
                if temp == self.head:
                    self.head = temp.next
                temp = temp.next
            else:
                temp = temp.next

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()

# Example usage
dll = DoublyLinkedList()
dll.insert_at_end('a')
dll.insert_at_end('b')
dll.insert_at_end('c')
dll.insert_at_end('b')
dll.insert_at_end('d')
dll.display()  # Output: a b c b d
dll.remove_character('b')
dll.display()  # Output: a c d
```

```python
#8. WAP to divide a doubly linked list from middle and rejoin them in reverse manner
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node
            new_node.prev = temp

    def reverse_list(self, head):
        temp = None
        current = head
        while current:
            temp = current.prev
            current.prev = current.next
            current.next = temp
            current = current.prev
        if temp:
            return temp.prev
        return head

    def find_middle(self):
        slow = self.head
        fast = self.head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        return slow

    def divide_and_reverse(self):
        if not self.head or not self.head.next:
            return
        mid = self.find_middle()

        # Split the list into two halves
        second_half = mid.next
        mid.next = None
        if second_half:
            second_half.prev = None

        # Reverse both halves
        first_half = self.reverse_list(self.head)
        second_half = self.reverse_list(second_half)

        # Rejoin the reversed lists
        temp = first_half
        while temp.next:
            temp = temp.next
        temp.next = second_half
        if second_half:
            second_half.prev = temp

        # Update head
        self.head = first_half

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()

# Example usage
dll = DoublyLinkedList()
dll.insert_at_end(1)
dll.insert_at_end(2)
dll.insert_at_end(3)
dll.insert_at_end(4)
dll.insert_at_end(5)
dll.insert_at_end(6)

print("Original List:")
dll.display()  # Output: 1 2 3 4 5 6

dll.divide_and_reverse()

print("After Dividing and Reversing:")
dll.display()  # Output: 3 2 1 6 5 4
```

```python
#9.WAP to reverse a doubly linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node
            new_node.prev = temp

    def reverse(self):
        temp = None
        current = self.head
        while current:
            temp = current.prev
            current.prev = current.next
            current.next = temp
            current = current.prev
        if temp:
            self.head = temp.prev

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()

# Example usage
dll = DoublyLinkedList()
dll.insert_at_end(1)
dll.insert_at_end(2)
dll.insert_at_end(3)
dll.insert_at_end(4)
dll.insert_at_end(5)

print("Original List:")
dll.display()  # Output: 1 2 3 4 5

dll.reverse()

print("Reversed List:")
dll.display()  # Output: 5 4 3 2 1
```

```python
#10. WAP to arrange a binary tree with a doubly linked list
class TreeNode:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

class DoublyLinkedListNode:
    def __init__(self, key):
        self.val = key
        self.prev = None
        self.next = None

class BinaryTreeToDoublyLinkedList:
    def __init__(self):
        self.head = None
        self.prev = None

    def binary_tree_to_dll(self, root):
        if root is None:
            return

        # Convert left subtree
        self.binary_tree_to_dll(root.left)

        # Convert current node
        if self.prev is None:
            self.head = DoublyLinkedListNode(root.val)
            self.prev = self.head
        else:
            current = DoublyLinkedListNode(root.val)
            self.prev.next = current
            current.prev = self.prev
            self.prev = current

        # Convert right subtree
        self.binary_tree_to_dll(root.right)

    def display(self):
        temp = self.head
        while temp:
            print(temp.val, end=" ")
            temp = temp.next
        print()

# Example usage
root = TreeNode(4)
root.left = TreeNode(2)
root.right = TreeNode(6)
root.left.left = TreeNode(1)
root.left.right = TreeNode(3)
root.right.left = TreeNode(5)
root.right.right = TreeNode(7)

converter = BinaryTreeToDoublyLinkedList()
converter.binary_tree_to_dll(root)
print("Doubly Linked List from Binary Tree:")
converter.display()  # Output: 1 2 3 4 5 6 7
```
